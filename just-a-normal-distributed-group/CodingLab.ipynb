```python
import pandas as pd
import glob
from collections import Counter

def clean_on_sitc(df):
    new_df = df[["DECLARANT_ISO", "PARTNER_ISO","PRODUCT_SITC", "FLOW","QUANTITY_IN_KG","VALUE_IN_EUROS"
                ,"PERIOD"]]
    new_df = new_df[new_df["PRODUCT_SITC"] >= "57000"]
    new_df = new_df[new_df["PRODUCT_SITC"] < "59000"]
    return new_df


list_of_eu_countries = ["AT","BE","BG","CY","CZ","DE","DK","EE","ES","FI","FR","GB","GR","HR","HU","IE",
        "IT","LT","LU","LV","MT","NL","PL","PT","QR","QV","QY","RO","SE","SI","SK"]   
```

```python
final_output = pd.DataFrame(columns=["Period", "Export_Value_in_Euros", "Export_Quantity_in_KG", 
           "Import_Value_in_Euros", "Import_Quantity_in_KG"])
```

```python
dir_path = "/home/cratus/Project/Data"
file_path = "/home/cratus/Project/Data/full202002.dat"
all_data = glob.glob(dir_path + "/*.dat")
```

```python
for file in all_data:
    df = pd.read_csv(file, index_col=None, header=0)
    df1 = clean_on_sitc(df)
    for country in list_of_eu_countries:
        df1.drop(df1[df1['PARTNER_ISO'] == country].index, inplace = True)
    
    period = df["PERIOD"].iloc[2]
    
    monthly_import = df1.copy()
    monthly_import.drop(monthly_import[monthly_import["FLOW"] == 2].index, inplace = True)
    
    monthly_import_quantity = monthly_import["QUANTITY_IN_KG"].sum()
    monthly_import_value = monthly_import["VALUE_IN_EUROS"].sum()
    
    monthly_export = df1.copy()
    monthly_export.drop(monthly_export[monthly_export["FLOW"] == 1].index, inplace = True)
    
    monthly_export_quantity = monthly_export["QUANTITY_IN_KG"].sum()
    monthly_export_value = monthly_export["VALUE_IN_EUROS"].sum()
    
    final_output.loc[len(final_output.index)] = [period, monthly_export_value, monthly_export_quantity
                                                 , monthly_import_value, monthly_import_quantity]
    print(f"Finished {period}")
```
