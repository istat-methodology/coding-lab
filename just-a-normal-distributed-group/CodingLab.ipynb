```python
import pandas as pd
import glob
from collections import Counter

def clean_on_sitc(df):
    new_df = df[["DECLARANT_ISO", "PARTNER_ISO","PRODUCT_SITC", "FLOW","QUANTITY_IN_KG","VALUE_IN_EUROS"
                ,"PERIOD"]]
    new_df = new_df[new_df["PRODUCT_SITC"] >= "57000"]
    new_df = new_df[new_df["PRODUCT_SITC"] < "59000"]
    return new_df
```

```python
dir_path = "/home/cratus/Project/data"
file_path = "/home/cratus/Project/data/full202002.dat"
all_data = glob.glob(dir_path + "/*.dat")
```

```python
list_of_eu_countries = ["AT","BE","BG","CY","CZ","DE","DK","EE","ES","FI","FR","GB","GR","HR","HU","IE",
"IT","LT","LU","LV","MT","NL","PL","PT","QR","QV","QY","RO","SE","SI","SK"]
```

```python
eu=pd.read_csv("https://ec.europa.eu/eurostat/estat-navtree-portlet-prod/BulkDownloadListing?sort=1&file=comext%2FCOMEXT_METADATA%2FCLASSIFICATIONS_AND_RELATIONS%2FENGLISH%2FGEOZONES-PARTNERS_ISO.txt",sep="\t",header=None)[[0,1]].rename({0:'DECLARANT_ISO_CODE', 1:'DECLARANT_ISO'}, axis=1)
eu=eu[(eu.DECLARANT_ISO_CODE==1010)]
eu.drop_duplicates(inplace=True)

# Deleted the following DECLARANT_ISOs: QR, QV, QY
eu = eu.drop([353,355, 356],axis="index")
```

```python
for file in all_data:
    df = pd.read_csv(file, index_col=None, header=0)
    df = df.merge(eu, on = "DECLARANT_ISO")
    result = clean_on_sitc(df)
    result.to_excel(f"{file}.xlsx")
```
